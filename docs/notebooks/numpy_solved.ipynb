````python
colores = ("rojo","amarillo","verde","purpura","azul")

def misegunda_funcion(tup):
    """
    Esta rutina imprime cada uno de los input de una tupla y
    retorna la cantidad de elementos
    
    Parámetros
    ----------
    tup : tupla
        Representa los elementos que queremos imprimir
    
    Retorna
    ---------
    cant : float
        Cantidad de elementos impresos
    """
    # borrar
    for i in tup:
        print(i)
    
    return len(tup)

a = misegunda_funcion(colores)
print("El tamaño de la lista es:{}".format(a))
````

## Aplicación: Conjunto de Mandelbrot 

- poner código aquí para crear una matriz booleana que contenga True si un punto pertenece al conjunto de Mandelbrot

- valores razonables: 50 iteraciones, umbral a 100, y una cuadrícula de 300x300

- Escoger los parámetros de acuerdo a lo que ustedes consideren adecuado

````python
npts = 3000j
y, x = np.ogrid[-1.5:1.5:npts, -1.5:1.5:npts]
c = x+1j*y
z = c
for n in range(50):
    z = np.exp(z**2)+c #This is an example of how we can play with this functions
imdata = (np.abs(z) < 50)
print(imdata.shape)
plt.figure(figsize=(10,10))
plt.imshow(imdata, cmap='bone')
plt.savefig(fname="mandelbrot.pdf",dpi=200)
````

## Problema 3: Caminata aleatoria

````python
length_of_walk = 10000
realizations = 15
angles = 2*np.pi*np.random.rand(length_of_walk, realizations)
x = np.cumsum(np.cos(angles), axis=0)
y = np.cumsum(np.sin(angles), axis=0)
plt.plot(x, y)
plt.axis('scaled')
x.shape
````
